```python
import random

def jugar():
    opciones = ["piedra", "papel", "tijera", "lagarto", "spock", "fuego", "agua", "aire"]

    reglas = {
        "tijera": {"papel": "corta", "lagarto": "decapita", "aire": "rompe"},
        "papel": {"piedra": "cubre", "spock": "refuta", "agua": "absorbe"},
        "piedra": {"tijera": "aplasta", "lagarto": "aplasta", "fuego": "soporta"},
        "lagarto": {"spock": "envenena", "papel": "devora", "aire": "inhala"},
        "spock": {"tijera": "rompe", "piedra": "vaporiza", "fuego": "extingue"},
        "fuego": {"papel": "quema", "tijera": "derrite", "aire": "consume"},
        "agua": {"fuego": "apaga", "papel": "moja", "piedra": "erosiona"},
        "aire": {"agua": "evapora", "lagarto": "sofoca", "fuego": "aviva"}
    }

    puntuacion = {"usuario": 0, "computadora": 0, "empates": 0}
    historial = []

    print("üïπÔ∏è Bienvenido a la versi√≥n EXTENDIDA de 'Piedra, Papel, Tijera, Lagarto, Spock + elementos'")
    print("Opciones disponibles:", ", ".join(opciones))
    print("Escribe una opci√≥n o usa 'salir', 'historial' o 'reiniciar'\n")

    while True:
        eleccion_usuario = input("Tu elecci√≥n: ").lower().strip()

        if eleccion_usuario == "salir":
            print("\nüõë Juego terminado. Gracias por jugar.")
            print(f"üèÅ Resultado final: T√∫: {puntuacion['usuario']} | Computadora: {puntuacion['computadora']} | Empates: {puntuacion['empates']}")
            break

        if eleccion_usuario == "historial":
            if historial:
                print("\nüìú Historial de partidas:")
                for i, jugada in enumerate(historial, 1):
                    print(f"{i}. T√∫: {jugada['usuario']} | PC: {jugada['computadora']} -> {jugada['resultado']}")
            else:
                print("\nüîç No hay jugadas registradas a√∫n.")
            continue

        if eleccion_usuario == "reiniciar":
            puntuacion = {"usuario": 0, "computadora": 0, "empates": 0}
            historial.clear()
            print("\nüîÑ Juego reiniciado. ¬°Vamos de nuevo!")
            continue

        if eleccion_usuario not in opciones:
            print("‚ùå Opci√≥n inv√°lida. Intenta de nuevo.")
            continue

        eleccion_computadora = random.choice(opciones)
        print(f"La computadora eligi√≥: {eleccion_computadora}")

        if eleccion_usuario == eleccion_computadora:
            print("ü§ù ¬°Empate!")
            puntuacion["empates"] += 1
            historial.append({
                "usuario": eleccion_usuario,
                "computadora": eleccion_computadora,
                "resultado": "Empate"
            })
        elif eleccion_computadora in reglas.get(eleccion_usuario, {}):
            accion = reglas[eleccion_usuario][eleccion_computadora]
            print(f"‚úÖ ¬°T√∫ ganas! {eleccion_usuario.capitalize()} {accion} a {eleccion_computadora}.")
            puntuacion["usuario"] += 1
            historial.append({
                "usuario": eleccion_usuario,
                "computadora": eleccion_computadora,
                "resultado": "Ganaste"
            })
        elif eleccion_usuario in reglas.get(eleccion_computadora, {}):
            accion = reglas[eleccion_computadora][eleccion_usuario]
            print(f"‚ùå ¬°T√∫ pierdes! {eleccion_computadora.capitalize()} {accion} a {eleccion_usuario}.")
            puntuacion["computadora"] += 1
            historial.append({
                "usuario": eleccion_usuario,
                "computadora": eleccion_computadora,
                "resultado": "Perdiste"
            })
        else:
            print("‚ö†Ô∏è No hay reglas definidas entre estas dos opciones. Se considera empate.")
            puntuacion["empates"] += 1
            historial.append({
                "usuario": eleccion_usuario,
                "computadora": eleccion_computadora,
                "resultado": "Empate"
            })

        print(f"Marcador: T√∫: {puntuacion['usuario']} | Computadora: {puntuacion['computadora']} | Empates: {puntuacion['empates']}\n")

# Ejecutar el juego
jugar()
```
